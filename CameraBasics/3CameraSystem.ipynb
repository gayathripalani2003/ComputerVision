```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ed = 0.2 #Error in disparity.
zrange = np.arange(0.01,5.0,0.5) #Depth range is upto 5.
facallength = 150
fig,axes = plt.subplots(2,1)

# Baseline is 1 for camera 1 and 2.
baseline = 0.8
ez1 = [(z*z*ed)/(facallength*baseline) for z in zrange]
axes[0].plot(zrange,ez1)
axes[0].set(ylabel = "Error(m)")
axes[0].set_title(f"Base Line:{baseline}")

#Baseline is 2 for camera 1 and 3.
baseline = 0.4
ez2 = [(z*z*ed)/(facallength*baseline) for z in zrange]
axes[1].plot(zrange,ez2)
axes[1].set(ylabel="Error(m)")
axes[1].set_title(f"Base Line:{baseline}")




```
